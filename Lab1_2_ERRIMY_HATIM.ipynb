# Web Scraping

Le Web Scraping (grattage) est une technique utilisée par les développeurs pour extraire  (littéralement « gratter ») des informations d’un site Internet.  

# How Does Web Scraping Work?
1. <b>Request-Response: </b>
    la Lib requests permet de récupérer les pages Web via le protocole http(s)
2. <b>Parse and Extract: </b>
    la Lib BeautifulSoup va analyser ces pages et extraire les données 
3. <b>Manipulate and save Data: </b> la Lib pandas pour la manipulation et le stockage de ces données.

```python
# requests: make HTTP requests
# Beautiful: library for parsing HTML and XML documents; in order to extract data.
# Pandas: library for data manipulation.
import  requests
from  bs4 import BeautifulSoup
import  pandas as pd
page=requests.get('https://www.jumia.ma/telephones-smartphones/')
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
# find <div class="info">
phones = soup.find_all("div", {"class": "info"})
images = soup.find_all("img", {"class": "img"})
for i in phones:
    print(i.prettify())
for img in images:
    print(img.prettify())
```

```python
#items=phones.find_all(class_='info')
#print(items[0])
phone_name=[i.find(class_='name').get_text() for i in phones]
phone_price=[i.find(class_='prc').get_text() for i in phones]
phone_images=[img.get('data-src') for img in images]
#phone_reviews=[item.find(class_='rev').get_text() for item in items]

print(phone_name)
print(phone_price)
print(phone_images)
```
